```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("spreads.csv", parse_dates=["timestamp"])
df.head()
```

```python
plt.plot(df["timestamp"], df["spread"])
plt.xlabel("Time (UTC)")
plt.ylabel("Mid-price Spread")
plt.title("Binance – Kraken Spread Over Time")
plt.show()
```
